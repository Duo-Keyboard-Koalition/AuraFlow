```python
from langchain_ollama import OllamaEmbeddings
from langchain_pinecone import PineconeVectorStore
from pinecone import Pinecone
from uuid import uuid4
# import the import .env stuff
import os
from dotenv import load_dotenv
load_dotenv()
PINECONE_API_KEY = os.getenv("PINECONE_API_KEY")


class pineconeVectorModelLLAMA3Embed:
    def __init__(self, index_string, k = 1, score_threshold = 0.5):
        embeddings = OllamaEmbeddings(
            model="llama3",
        )
        pc = Pinecone(api_key=PINECONE_API_KEY)
        index = pc.Index(index_string)
        self.embeddings = embeddings
        vector_store = PineconeVectorStore(index=index, embedding=embeddings)
        self.vector_store = vector_store
        self.retriever = vector_store.as_retriever(
            search_type="similarity_score_threshold",
            search_kwargs={"k": k, "score_threshold": score_threshold},
        )
    def get_retriever(self):
        return self.retriever
    '''
    Takes in a list of documents
    [
    Document(
        page_content="Building an exciting new project with LangChain - come check it out!",
        metadata={"source": "tweet"},
    ),
    ....
    ]
    '''
    def add_documents(self, documents, ids = None):
        if ids is None:
            ids = [str(uuid4()) for _ in range(len(documents))]
        return self.vector_store.add_documents(documents=documents, ids=ids)




influencers_db = pineconeVectorModelLLAMA3Embed("meta3-hackathon")

from uuid import uuid4
from langchain_core.documents import Document

# Documents tagged with different influencers and their vibes
document_1 = Document(
    page_content="LuminousLisa embodies radiance, positivity, and warmth, inspiring her audience with an aura of optimism. She aligns beautifully with SolarSync, a brand dedicated to renewable energy and innovation, symbolizing a brighter and sustainable future.",
    metadata={"influencer": "LuminousLisa", "vibe": ["Radiance", "Positivity", "Warmth"], "brand": "SolarSync", "identity": "Renewable energy and innovation"}
)

document_2 = Document(
    page_content="EdgyEthan exudes boldness, rebellion, and courage, appealing to those who dare to defy conventions. His dynamic persona resonates perfectly with UrbanEdge, a brand representing street style and attitude, empowering individuals to make bold fashion statements.",
    metadata={"influencer": "EdgyEthan", "vibe": ["Boldness", "Rebellion", "Courage"], "brand": "UrbanEdge", "identity": "Street style and attitude"}
)

document_3 = Document(
    page_content="ChillCharlie is the epitome of calm, authenticity, and grounded energy, connecting with audiences seeking tranquility and genuine interactions. His vibe harmonizes seamlessly with ZenFlow, a brand rooted in mindfulness and serenity, offering moments of peace in a busy world.",
    metadata={"influencer": "ChillCharlie", "vibe": ["Calm", "Authenticity", "Grounded"], "brand": "ZenFlow", "identity": "Mindfulness and serenity"}
)

document_4 = Document(
    page_content="VibrantVictoria channels energy, creativity, and passion, captivating her audience with a zest for life and artistic expression. Her infectious enthusiasm complements ColorBurst, a brand focused on art and self-expression, inspiring creativity through bold and colorful designs.",
    metadata={"influencer": "VibrantVictoria", "vibe": ["Energy", "Creativity", "Passion"], "brand": "ColorBurst", "identity": "Art and self-expression"}
)

document_5 = Document(
    page_content="SophisticatedSam personifies elegance, luxury, and refinement, appealing to audiences with a taste for prestige and exclusivity. His refined persona aligns effortlessly with Grandeur, a brand that defines prestige and luxury, creating timeless impressions of sophistication.",
    metadata={"influencer": "SophisticatedSam", "vibe": ["Elegance", "Luxury", "Refinement"], "brand": "Grandeur", "identity": "Prestige and exclusivity"}
)

# List of documents
documents = [
    document_1,
    document_2,
    document_3,
    document_4,
    document_5,
]

# Assign UUIDs to each document
uuids = [str(uuid4()) for _ in range(len(documents))]

# Add documents to the vector store
influencers_db.add_documents(documents=documents, ids=uuids)

print("Influencer-brand connections successfully added to AuraMatch!")
```
